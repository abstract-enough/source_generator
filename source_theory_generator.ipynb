```python
def partition_list(n):
  operator_list = [bin(2**(n-1)-i-1)[2:].zfill(n-1) for i in range(2**(n-1))]
  partition_list = []

  for i in range(2**(n-1)):
    partition = []
    if operator_list[i][0] == "1":
      partition.append(1)
      last_plus = 0
    else:
      last_plus = -1
    for j in range(1,n-1):
      if operator_list[i][j] == "1":
        partition.append(j-last_plus)
        last_plus = j
    partition.append(n-1-last_plus)

    partition_list.append(partition)

  return partition_list
```

```python
def short_partition_list(partition_list):
  duplicates = True
  while duplicates:
    duplicates = False
    for i in range(len(partition_list)):
      for j in range(i+1, len(partition_list)):
        partition_1 = partition_list[i]
        partition_2 = partition_list[j]
        partition_1.sort()
        partition_2.sort()
        if str(partition_1) == str(partition_2):
          partition_list.pop(j)
          duplicates = True
          break
      if duplicates:
        break
  return partition_list[:-1]
```

```python
import ast

def minimal_sources(n):

  sources = short_partition_list(partition_list(n))

  hit = True
  while hit:
    hit = False

    #print(sources)
    for i in range(len(sources)):

      source_string = str(sources[i]).replace(" ","")


      for j in range(len(source_string)):
        char1 = source_string[j]
        if char1 == "[" or char1 == ",":
          char2 = source_string[j+1]
          if char2 != "[":
            k = find_end_number(source_string, j)
            new_partitions, hit = replace_part(source_string, j, k, hit)
            if hit:
              break
      if hit:
        literal = ast.literal_eval("["+new_partitions+"]")
        #print("l",literal)

        sources = sources[:i]+literal+sources[(i+1):]
        break

  return sources


def replace_part(source_string, i, j, hit):

  number = int(source_string[(i+1):j])
  if number>=2:
    hit=True

    source_literal = ast.literal_eval(source_string)
    replace_partitions = short_partition_list(partition_list(number))
    new_sources = []

    #print(source_string)
    for partition in replace_partitions:
      #print(source_string[:(i+1)])
      #print(str(partition)[1:-1])
      #print(source_string[j:])
      new_sources.append(\
        ast.literal_eval(\
        source_string[:(i+1)]+\
          str(partition)+\
          source_string[j:]))

    source_string = str(new_sources)[1:-1]
    #print("new",number,source_string)

  return source_string, hit


def find_end_number(partitions,i):
  for j in range(i+1,len(partitions)):
    if partitions[j] == "," or partitions[j] == "]":
      return j
```
