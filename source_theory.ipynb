```python
def partition_set(number):
  answer = set()
  answer.add((number, ))
  for x in range(1, number):
    for y in partition_set(number - x):
      answer.add(tuple(sorted((x, ) + y)))
  return answer
```

```python
def short_partition_string(n):
  partitions = partition_set(n)
  partitions_list = []
  for partition_tuple in partitions:
    if len(partition_tuple) >=2:
      partition_list = []
      for element in partition_tuple:
        partition_list.append(element)
      partitions_list.append(partition_list)
  return str(partitions_list).replace(" ","")
```

```python
def sources_equal(source1, source2):

  if source1 == 1 and source2==1:
    return True
  if source1 == 1 or source2 == 1:
    return False

  found = True
  while(found):
    found=False
    subcollect1=source1[0]
    for j in range(len(source2)):
      subcollect2 = source2[j]
      if sources_equal(subcollect1, subcollect2):
        source1 = source1[1:]
        source2 = source2[:j] + \
                  source2[(j+1):]
        found = True
        if len(source1)==0 and len(source2)==0:
          return True
        if len(source1)==0 or len(source2)==0:
          return False
        break
  return False
```

```python
def unique_source_list(source_list):

  duplicates = True
  while(duplicates):
    duplicates = False
    for i in range(len(source_list)-1):
      for j in range(i+1, len(source_list)):
        if sources_equal(source_list[i], source_list[j]):
          source_list.pop(j)
          duplicates = True
          break
      if duplicates == True:
        break

  return source_list
```
